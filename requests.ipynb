```python
import requests

def get_lat_long(address, api_key):
    # Endpoint da API de Geocoding
    url = "https://maps.googleapis.com/maps/api/geocode/json"
    
    # Parâmetros da requisição
    params = {
        "address": address,
        "key": api_key
    }
    
    try:
        response = requests.get(url, params=params)
        response.raise_for_status()  # Verifica se houve erro na requisição HTTP
        
        data = response.json()
        
        if data["status"] == "OK":
            # Extraindo latitude e longitude do primeiro resultado
            location = data["results"][0]["geometry"]["location"]
            lat = location["lat"]
            lng = location["lng"]
            return lat, lng
        else:
            return f"Erro na API: {data['status']}", None
            
    except requests.exceptions.RequestException as e:
        return f"Erro de conexão: {e}", None

# Configurações
SUA_CHAVE_API = "SUA_CHAVE_AQUI"
endereco = "Av. Paulista, 1000, São Paulo, SP"

latitude, longitude = get_lat_long(endereco, SUA_CHAVE_API)

if longitude:
    print(f"Endereço: {endereco}")
    print(f"Latitude: {latitude}")
    print(f"Longitude: {longitude}")
else:
    print(latitude)
```
